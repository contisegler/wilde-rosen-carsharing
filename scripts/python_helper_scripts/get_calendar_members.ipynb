# Get Calendar Members

## How to get an access token:

1. Go to **OAuth 2.0 Playground**: https://developers.google.com/oauthplayground/
2. In **Step 1**, find and select: `https://www.googleapis.com/auth/calendar.acls.readonly`
3. Click **Authorize APIs** and log in with your Google account
4. In **Step 2**, click **Exchange authorization code for tokens**
5. Copy the **Access token** and paste it below in the `access_token` variable

**Note**: Token expires in 1 hour

```python
import requests
from google.cloud import firestore
```

```python
access_token = ''
calendar_id = ''
```

```python
url = f'https://www.googleapis.com/calendar/v3/calendars/{calendar_id}/acl'
headers = {'Authorization': f'Bearer {access_token}'}
response = requests.get(url, headers=headers)
acl_data = response.json()

email_addresses = [rule['scope']['value'] for rule in acl_data.get('items', []) if rule['scope']['type'] == 'user']
```

```python
db = firestore.Client()

for email in email_addresses:
    account_ref = db.collection('accounts').document()
    account_ref.set({
        'emails': [email]
    })
```
